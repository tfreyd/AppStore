the "english" function does not handle well when there are unsual characters or slimeys. the function will hence be less sensitive and 3 

```python
def english(name):
    aberation=0
    
    for character in name:      
        if ord(character)>127:
            aberation += 1
            
    if aberation>3:
        return False                    
    return True

def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_english('Instachat üòú'))
```

```python
def freq_table(dataset, index):
    #building of the dictionnary
    content_column={}
       
    for row in dataset:
        string=row[index]
        if string in content_column:
            content_column[string] += 1
        else:
            content_column[string] = 1
#buiding of the frequencies table                
    total=len(dataset)
    freq_table={}
    
    for string in content_column:
        freq_table[string]= content_column[string]/total *100
    return freq_table

#in order to sort the frequencies table in reverse order
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
        
```

```python
#frequency table of prime_genre in iOS 
#and the average number of users per genre
popular_genre={}
most_popular=0
max_avg_users=0
for genre in freq_table(ios_final,11): #return the frequence table of the index in the dataset
    total=0
    len_genre=0
   
    
    for row in ios_final:
        genre_app=row[11]
        if genre_app == genre:
            n_user=int(row[5])
            total += n_user
            len_genre += 1
    avg_users=total/len_genre
    popular_genre[genre]=avg_users
    if avg_users > max_avg_users:
        max_avg_users=avg_users
        most_popular=genre
    
    print(int(total/len_genre)," average users", ' for', genre)
     
print('The most popular genre in iOS is ',most_popular,' with ',int(max_avg_users)," users")         
            
            
```
